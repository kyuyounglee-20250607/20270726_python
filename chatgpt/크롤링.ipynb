```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

options = Options()
# options.add_argument("--headless")  # 일단 꺼보세요
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://news.daum.net/")
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "/html/body/div[2]/div[1]/div[1]/article"))
)
time.sleep(2)

html = driver.page_source
print("✅ HTML 길이:", len(html))
print("✅ 'article' 포함 여부:", "article" in html)

driver.quit()
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")
articles = soup.find_all("article")

print(f"🔍 전체 article 태그 수: {len(articles)}")
for i, a in enumerate(articles[:3]):
    print(f"{i+1}. 내용 일부: {a.get_text(strip=True)[:300]}")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")

# 1. 뉴스 카드 리스트 선택
news_items = soup.select("ul.list_newsheadline2 > li > a.item_newsheadline2")

print(f"📰 뉴스 카드 수: {len(news_items)}")

# 2. 각 뉴스 카드에서 제목, 링크, 요약 추출
for i, item in enumerate(news_items[:5]):  # 예시로 5개만
    link = item.get("href")
    title_tag = item.select_one("strong.tit_txt")
    desc_tag = item.select_one("p.desc_txt")

    title = title_tag.get_text(strip=True) if title_tag else "제목 없음"
    desc = desc_tag.get_text(strip=True) if desc_tag else "요약 없음"

    print(f"{i+1}. 📰 제목: {title}")
    print(f"   🔗 링크: {link}")
    print(f"   📄 요약: {desc}")
    print("-" * 80)
```
