```python
import requests
from bs4 import BeautifulSoup

# 테스트할 URL
url = "https://news.naver.com/section/105"

# User-Agent 포함 헤더
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/139.0.0.0 Safari/537.36",
    "Referer": "https://www.naver.com/",
    "Accept-Language": "ko-KR,ko;q=0.9"
}

res = requests.get(url, headers=headers)

# BeautifulSoup으로 파싱
soup = BeautifulSoup(res.text, "html.parser")

# 모든 <a> 태그 찾기
a_tags = soup.find_all("a")

print(f"총 {len(a_tags)}개의 <a> 태그가 발견되었습니다.\n")

# 앞 20개만 출력
for a in a_tags[:20]:
    href = a.get("href")
    text = a.get_text(strip=True)
    print(f"텍스트: {text} | 링크: {href}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# -----------------------------
# 1. Chrome 옵션 설정 (헤드리스 모드)
# -----------------------------
options = Options()
options.add_argument("--headless")  # 브라우저 창을 안 띄움
options.add_argument("--disable-gpu")
options.add_argument("--window-size=1920,1080")  # 일부 사이트에서 필요

driver = webdriver.Chrome(options=options)

# -----------------------------
# 2. 네이버 뉴스 IT/과학 섹션 접속
# -----------------------------
url = "https://news.naver.com/section/105"
driver.get(url)

# -----------------------------
# 3. 기사 로딩 대기 (최대 10초)
# -----------------------------
wait = WebDriverWait(driver, 10)
articles = wait.until(EC.presence_of_all_elements_located(
    (By.CSS_SELECTOR, "div.sa_text a.sa_text_title")  # 기사 제목 CSS 선택자
))

# -----------------------------
# 4. 기사 제목 + 링크 출력
# -----------------------------
print(f"총 {len(articles)}개의 기사 추출됨\n")
for article in articles:
    title = article.text.strip()
    link = article.get_attribute("href")
    print(f"제목: {title}")
    print(f"링크: {link}")
    print("-" * 50)

# -----------------------------
# 5. 브라우저 종료
# -----------------------------
driver.quit()
```
